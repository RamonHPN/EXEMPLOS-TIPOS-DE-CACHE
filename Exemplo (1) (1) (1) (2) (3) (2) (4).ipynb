5.2 - Cáculo da quantidade de bits necessários para uma determinada memória cache. Considere um sistema de computção com uma memória cache de 32 KB de capacidade, constituída de linhas com 8 bytes de largura. a MP possuiuma capacidade de 16 MB. 

> R = Para se saber o o total de bits necessários é preciso somar os bits de dados e os bits usados na coluna tag.

> bits de dados = quantidade de linhas x largura em bits de cada linha.

> total de bits de tag = quantidade de linhas x largura do campo tag.


```python
def TOTAL_DE_BITS_DADOS(Q_linhas, Largura_bits):
    x = Q_linhas * Largura_bits
    return x

a = int(input(print(" Digite a quantidade de linhas em bits: ")))
b = int(input(print(" Digite a largura em bits da linha: ")))
print("total de bits de dados =",TOTAL_DE_BITS_DADOS(a, b), "bits")
```

```python
def QUANT_LINHAS(capacidade, Largura_linhas):
    x = capacidade / Largura_linhas
    return x

a = int(input(print(" Digite a capacidade cache: ")))
b = int(input(print(" Digite a largura em bits da linha: ")))
print("quantidade de linhas =",QUANT_LINHAS(a, b), "bytes")
```

> Largura do campo tag = quant de blocos / quant de linhas

```python
def QUANT_BLOCOS(capacidadeMP, Largura_linha):
    x = capacidadeMP / Largura_linha
    return x

a = int(input(print(" Digite a capacidade MP: ")))
b = int(input(print(" Digite a largura em bits da linha: ")))
print("quantidade de linhas =",QUANT_BLOCOS(a, b), "Mbytes")
```

> QUANT_BLOCOS = 2 MB = 2**21 = 2097152

> QUANT_LINHAS = 4KB = 2**12 = 4096

```python
def QUANT_BLOCOS_POR_LINHA(QUANT_BLOCOS, QUANT_LINHAS):
    x = QUANT_BLOCOS / QUANT_LINHAS
    return x

a = int(input(print(" Digite a Quantidade de blocos ")))
b = int(input(print(" Digite a Quantidade de linhas ")))
print("quantidade de blocos por linhas =", QUANT_BLOCOS_POR_LINHA(a,b),)
```

```python
import math
def campo_TAG(N):
    E = math.log( N, 2)
    return E
a = int(input(print(" Digite a Quantidade de blocos por linha: ")))
print("Campo TAG =",campo_TAG(a))
```

> QUANTIDADE DE LINHAS = 4KB campo TAG = 9 bits

```python
def TOTAL_BITS_TAG(quant_linhas,campotag):
    x = quant_linhas * campotag
    return x

a = int(input(print(" Digite a Quantidadede linhas ")))
b = int(input(print(" Digite o campo tag ")))
print("total de bits dos tags =", TOTAL_BITS_TAG(a,b),)
```

> total de bits cache = total bits de dados + total bits dos tags tag

```python
def TOTAL_BITS_CACHE(TD,TT):
    x = TD + TT
    return x

a = int(input(print(" Digite o total bits de dados ")))
b = int(input(print(" Digite o total de bits da tag ")))
print("total de bits da cache =", TOTAL_BITS_CACHE(a,b))
```

5.7 - Seja uma MP constituída de blocos com largura de 32 bytes, associada a uma cache com 64KB. Em dado instante o processador realiza um acesso, colocando o seguinte endereço (expresso em algarismos hexadecimais): 3FC92B6. Determine qual deverá ser o valor binário do campo bloco que será localizado pelo sistema de controle da cache.

>largura blocos = 32*8

>capacidade cache = 64*(2**13)

```python
import math
largblocos = 32*8
capacache = 64*2**13

E = len("3FC92B6")*4
bytecamp = math.log2(largblocos/8)
qtdbloco = 2**E/2**bytecamp
campbloco = math.log2(qtdbloco)

inteiro = int("3FC92B6", base=16)
bi = bin(inteiro)[::-1]

#bytes
byte = ""
for i in range(int(bytecamp)):
    byte += bi[i]
byte = byte[::-1]
print("Bytes",byte,"bits")

#bloco
bloco = ""
for j in range(int(campbloco)):
    bloco += bi[j+int(bytecamp)]
bloco = bloco[::-1]
print("Blocos:",bloco,"bits")
```

> FIM

5.10 - Seja uma MP constituída de blocos com largura de 32 bytes, associada a uma cache com 64KB; a cache usa mapeamento por conjunto de 4. Em dado instante o processador realiza um acesso, colocando o seguinte endereço (expresso em algarismos hexadecimais): 3FC92B6. Determine qual deverá ser o valor binário do conjunto que será localizado pelo sistema de controle da cache.

>largura do bloco = 32*8 = 256

>capacidade cache = 64*(2**13) = 524288

>conjunto = 4

```python
largbloco = 32*8
capacache = 64*2**13
conjunto = 4

E = len("3FC92B6")*4
bytecamp = math.log2(largbloco/8)
print("Bytes:",int(bytecamp),"bits")

qtdlinhas = capacache/(2**bytecamp*8)

qtdconjunto=qtdlinhas/conjunto

conjuntocamp = math.log2(qtdconjunto)
print("Conjunto:",int(conjuntocamp),"bits")

tagcamp = E - conjuntocamp - bytecamp
print("TAG:",int(tagcamp),"bits")

print("Conjunto: 010010101")
print("tag: 0b111111110010")
print("O endereço da linha desejada é: 010010101")
```
